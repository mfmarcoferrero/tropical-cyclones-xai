```python
import netCDF4 as nc
ds_press = nc.Dataset('./nc_original_datasets/msl_pressure.nc')

press_arr = ds_press['msl'][:,0,:,:]
latitude_arr = ds_press['latitude'][:]
longitude_arr = ds_press['longitude'][:]

print(press_arr.shape)
print(latitude_arr.shape)
print(longitude_arr.shape)
```

```python
import numpy as np
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming you have a 3D numpy array with shape (N_samples, latitudes, longitudes)
# Replace this with your actual data
N_samples = 1000
latitudes  = latitude_arr[:N_samples].data
longitudes = longitude_arr[:N_samples].data
data_3d    = press_arr[:N_samples, :, :].data
```

```python
print(latitudes.shape)
print(longitudes.shape)
print(data_3d.shape)
```

```python
print(np.tile(np.repeat(latitudes, len(longitudes)), N_samples).shape)
print(np.tile(np.tile(longitudes, len(latitudes)),   N_samples).shape)
print(data_3d.flatten().shape)
```

```python
# Reshape the 3D array to (N_samples * latitudes * longitudes, 3)
data_3d_reshape = np.column_stack([
    np.tile(np.repeat(latitudes, len(longitudes)), N_samples),
    np.tile(np.tile(longitudes, len(latitudes)),   N_samples),
    data_3d.flatten()
])

# Create a KDE model for the 3D data
kde_3d = KernelDensity(bandwidth=1.0, kernel='gaussian')
kde_3d.fit(data_3d_reshape)
```

```python
print(data_3d_reshape[1000])
```

```python
N = 100
print('MIN P:', np.min(data_3d))
print('MAX P:', np.max(data_3d))
pressure_values = np.linspace(np.min(data_3d), np.max(data_3d), N)
for elem in pressure_values:
  print(kde_3d.score([(-15,50,elem)]))
```

```python
print(kde_3d.score([(-15,50,96789)]))
```

```python
N = 1000
# Create a grid of latitude, longitude, and pressure values for visualization
latitude_values = np.linspace(0, -30, 121)
longitude_values = np.linspace(30, 90, 241)
pressure_values = np.linspace(np.min(data_3d), np.max(data_3d), N)

# Create a meshgrid for latitude, longitude, and pressure
latitude_mesh, longitude_mesh, pressure_mesh = np.meshgrid(latitude_values, longitude_values, pressure_values)
print(latitude_mesh.shape, longitude_mesh.shape, pressure_mesh.shape)

# Flatten the meshgrid for evaluation
grid_3d = np.column_stack((latitude_mesh.ravel(), longitude_mesh.ravel(), pressure_mesh.ravel()))
print(grid_3d.shape)
```

```python
# Evaluate the log density on the grid for latitude, longitude, and pressure
log_density_3d = kde_3d.score_samples(grid_3d)
# Reshape the results to the original meshgrid shape
print(log_density_3d.shape)

density_3d = np.exp(log_density_3d).reshape(latitude_mesh.shape)
print(density_3d.shape)
```

```python
# Plot the results in 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_3d_reshape[:, 0], data_3d_reshape[:, 1], data_3d_reshape[:, 2], c='b', marker='o', alpha=0.5, label='Data Points')
ax.contourf(latitude_mesh, longitude_mesh, density_3d.sum(axis=2), zdir='z', offset=np.min(pressure_values), cmap='viridis', alpha=0.5)

ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Pressure')
ax.set_title('Kernel Density Estimation (KDE) in 3D')
ax.legend()

plt.show()
```

# Evaluate variables distribution over weeks of the year

```python
df = pd.read_csv('./DATASET_COMPLETED/4_zones_avg_std_10D.csv')

dates = df['DATE'].values
ydays = []
for date_str in dates:
    date = datetime.strptime(date_str, '%Y-%m-%d')
    day_of_year = date.timetuple().tm_yday
    ydays.append(day_of_year)
df['yday'] = ydays
df = df.set_index('DATE')

df_target = df['S.IndAll']
df_allvars = df.drop(['S.IndAll'], axis=1)

df_allvars
```

```python
df_allvars.columns[1::88]
```

```python
dates = df_allvars.index

def get_week_number(date_str):
    # Convert the date string to a datetime object
    date_object = datetime.strptime(date_str, '%Y-%m-%d')
    # Use strftime to get the ISO week number
    week_number = date_object.strftime('%V')
    return int(week_number)

week_number_list = []
for elem in dates:
  week_number_list.append(get_week_number(elem))

week_number_arr = np.array(week_number_list)
print(week_number_arr.shape)
print(week_number_arr)
```

```python
df_wind = df_allvars[['Wind_Gust_Mean_Z1-0', 'Wind_Gust_Mean_Z2-0', 'Wind_Gust_Mean_Z3-0', 'Wind_Gust_Mean_Z4-0']]
df_wind['WEEK'] = week_number_arr
df_wind
```

```python
wind_z1 = df_wind['Wind_Gust_Mean_Z1-0'].values
wind_z2 = df_wind['Wind_Gust_Mean_Z2-0'].values
wind_z3 = df_wind['Wind_Gust_Mean_Z3-0'].values
wind_z4 = df_wind['Wind_Gust_Mean_Z4-0'].values
z1_flag_arr = df_wind['WEEK'].values
z2_flag_arr = df_wind['WEEK'].values
z3_flag_arr = df_wind['WEEK'].values
z4_flag_arr = df_wind['WEEK'].values

wind_arr = np.concatenate((wind_z1, wind_z2, wind_z3, wind_z4))
flag_arr = np.concatenate((z1_flag_arr, z2_flag_arr, z3_flag_arr, z4_flag_arr))
# Create a DataFrame using the arrays
data = {'Wind': wind_arr, 'Week': flag_arr}
df_kde = pd.DataFrame(data)

df_kde
```

```python
# Create a 2D KDE plot
plt.figure(figsize=(12, 8))
sns.kdeplot(x='Wind', y='Week', data=df_kde, cmap="Greens", fill=True)

# Set custom y-axis tick values
#custom_y_ticks = [1, 2, 3, 4]
#plt.yticks(custom_y_ticks)

plt.title('2D Kernel Density Estimation - Wind and Week')
plt.xlabel('Wind Speed')
plt.ylabel('Week Number')

plt.show()
```
